```python
from langchain_core.tools import tool
from langchain_dartmouth.llms import ChatDartmouthCloud, ChatDartmouth


@tool
def multiply(a: int, b: int) -> int:
    """Multiply two numbers."""
    return a * b


model = ChatDartmouth()
model_with_tools = model.bind_tools([multiply])

model_with_tools.invoke("what's 42 x 7?")
```

```python
from langchain_core.tools import tool
from langgraph.prebuilt import create_react_agent
from langchain_dartmouth.llms import ChatDartmouthCloud, ChatDartmouth


@tool
def multiply(a: int, b: int) -> int:
    """Multiply two numbers."""
    return a * b


model = ChatDartmouthCloud()
agent = create_react_agent(model=model, tools=[multiply])
agent.invoke({"messages": [{"role": "user", "content": "what's 42 x 7?"}]})
```
